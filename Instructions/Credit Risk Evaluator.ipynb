```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df

```

```python
# Convert categorical data to numeric and separate target feature for training data
train_dummy_df = pd.get_dummies(train_df)
train_dummy_df
# train_dummy_df = train_dummy_df.drop(['Unnamed: 0', 'index'], axis=1)
# train_dummy_df
X_train = train_dummy_df.drop(["target_high_risk", "target_low_risk"], axis = 1).values
y_train = train_dummy_df.target_high_risk.values
```

```python
list(train_dummy_df.columns)
```

```python
# Convert categorical data to numeric and 
test_dummy_df = pd.get_dummies(test_df)
test_dummy_df
# test_dummy_df = test_dummy_df.drop(['Unnamed: 0', 'index'], axis=1)
# test_dummy_df
```

```python
list(test_dummy_df.columns)
```

```python
# add missing dummy variables to testing set+
train_cols = list(train_dummy_df.columns)
test_cols = list(test_dummy_df.columns)
cols_not_in_test = {c:0 for c in train_cols if c not in test_cols}
test_dummy = test_dummy_df.assign(**cols_not_in_test)
test_dummy
```

```python
# separate target feature for testing data
X_test = test_dummy.drop(["target_high_risk", "target_low_risk"], axis = 1).values
y_test = test_dummy.target_high_risk.values
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier

classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)
```

```python
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier = LogisticRegression()
classifier
classifier.fit(X_train_scaled, y_train)
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```
